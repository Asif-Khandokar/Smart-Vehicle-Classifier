Simple Class with a constructor and placeholder method

```python
#Creating the base class name vehicle.
class Vehicle:
  def __init__(self, vehicle_type, name, speed):
    self.vehicle_type = vehicle_type
    self.__name = name
    self.__speed = speed

#method for getting name and setting name
  @property
  def name(self):
    return self.__name
  @name.setter
  def name(self,name):
    self.__name = name

#method for setting speed and getting speed
  @property
  def speed(self):
    return self.__speed

  @speed.setter
  def speed(self, speed):
    self.__speed = speed

#Method for classifying the vehicle.

  def classify(self):
    return "High speed Vehicle" if self.speed > 100 else "Normal Vehicle"

#show information of parent

  def show_info(self):
    return (f"Type:{self.vehicle_type}, Name:{self.name},"
           f"Speed:{self.speed} km/h, Predicted;{self.classify}")

```

Inheritance

```python
class Car(Vehicle):
  def __init__(self, name, speed):
    super().__init__("Car",name, speed)

  def show_info(self):
    return (f"Type:{self.vehicle_type}, Name:{self.name}, "
             f"Speed:{self.speed}km/h, Predicted:{self.classify()}")

class Bike(Vehicle):
  def __init__(self,name,speed):
    super().__init__("Bike",name,speed)

  def show_info(self):
    return(f"Type:{self.vehicle_type}, Name:{self.name}, "
           f"Speed:{self.speed}km/h, Predicted:{self.classify()}")

```

factory function

```python
def create_vehicle(vehicle_type,name,speed):
  vt =vehicle_type.strip().lower()
  if vt == "car":
    return Car(name, speed)

  elif vt == "bike":
    return Bike(name, speed)

  else:
    raise ValueError(f"Unknown Vehicle Type: {vehicle_type}")
```

```python
def main_interactive():

    n = int(input("Number of vehicles").strip())
    vehicles = []

    for i in range(n):
        line = input(f"Enter vehicle #{i+1} (Type Name Speed): ").strip()
        parts = line.split()
        if len(parts) != 3:
            print("Please enter exactly 3 items: Type Name Speed (e.g., Car Toyota 150)")
            return
        vtype, name, speed_text = parts
        try:
            speed = int(speed_text)
        except ValueError:
            print("Speed must be an integer")
            return

        try:
            v = create_vehicle(vtype, name, speed)
        except Exception as e:
            print(f"Error: {e}")
            return

        vehicles.append(v)
        print(f"{v.vehicle_type} Added: {v.name}")

    print("\n--- Vehicle Details ---")
    for v in vehicles:
        print(v.show_info())


# Run the interactive main
main_interactive()
```
